```python
# create dictionary of indexes representing each location
# locations = set(list(relevant_data.location))
location_dic = {}
for i in range(len(relevant_data.index)):
    if relevant_data.location[i] not in location_dic:
        location_dic[relevant_data.location[i]] = [i]
    else:
        location_dic[relevant_data.location[i]].append(i)
       
# each entry is a list of all the indexes of that location.
```

```python
# converting the dates from strings d/m/y to datetime yyyy-mm-dd

import datetime

format_str = '%d/%m/%Y'
dates = list(relevant_data['date'].values)
for i in range(len(dates)):
    dates[i] = datetime.datetime.strptime(dates[i], format_str)
    dates[i] = dates[i].date()

relevant_data['date'] = dates
type(relevant_data.loc[0,'date'])
```
